```python
'''Use given Iris dataset and solve the following questions:
1.show the count of different types of species
2.find the max,m and mean in values of Sepallength, petallength, sepalwidth and petalwidth
draw a graph:
3.scatterplot(for sepallength and petallength)
4.bargraph
5.jointplot(for sepallength and sepalwidth)
6.facetgrid(for sepallength and sepalwidth for all species)
7.boxplot(for species and sepalwidth)
8.boxplot & stripplot(for sepallength and all species)
9.violinplot(for petallength and all species) 
10.pairplot(for all 4 columns with species)
'''
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.show the count of different types of species
data = pd.read_csv('iris.csv')
data.groupby('Species').size()
```

```python
#2.find the max,m and mean in values of Sepallength, petallength, sepalwidth and petalwidth draw a graph:
a = data.describe()
print(data.groupby('Species').mean())
print(data.groupby('Species').min())
print(data.groupby('Species').max())
```

```python
data = data.drop('Id', axis=1)
g = sns.pairplot(data, hue='Species')
plt.show()
```

```python
#3.scatterplot(for sepallength and petallength)
sns.set(style="ticks") 
plt.figure(figsize=(12,10))
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=data)
```

```python
#4.bargraph(for sepallength and petallength)
sns.set(style="ticks") 
plt.figure(figsize=(12,10))
sns.barplot(x='SepalLengthCm',y='PetalLengthCm',data=data)
```

```python
# 5.jointplot(for sepallength and sepalwidth)
sns.set(style="ticks") 
plt.figure(figsize=(12,10))
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=data)
```

```python
# 6.facetgrid(for sepallength and sepalwidth for all species)
sns.FacetGrid(data, hue="Species", height=5).map(sns.distplot, "SepalLengthCm").add_legend()
sns.FacetGrid(data, hue="Species", height=5).map(sns.distplot, "SepalWidthCm").add_legend()
```

```python
#7.boxplot(for species and sepalwidth
sns.boxplot(x='Species',y='SepalWidthCm',data=data)
```

```python
#8.boxplot & stripplot(for sepallength and all species)
sns.boxplot(x='Species',y='SepalLengthCm',data=data)
sns.stripplot(x='Species',y='SepalLengthCm',data=data)
```

```python
#9.violinplot(for petallength and all species)

sns.violinplot(x='Species',y='PetalLengthCm',data=data)
```

```python
# 10.pairplot(for all 4 columns with species)
sns.pairplot(data, hue="Species")
```
